```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Data/googleform.csv")
```

```python
df.head()
```

```python
# split Number of previous treks by - and keep the first value
df['Number of Previous treks?'] = df['Number of Previous treks?'].str.split('-').str[0]
# remove all non digit characters from Number of Previous treks? and then split the string into a list and keep only the first element
df['Number of Previous treks?'] = df['Number of Previous treks?'].str.replace(r'\D', '').str.split().str[0]
# remove all non digit characters from Height (in cm) , Weight (in kg)	and Age
df['Height (in cm)'] = df['Height (in cm)'].str.replace(r'\D', '')
df['Weight (in kg)'] = df['Weight (in kg)'].str.replace(r'\D', '')
# if Age contains - then split the string into a list and keep only the first element
df['Age'] = df['Age'].str.split('-').str[0]
df['Age'] = df['Age'].str.replace(r'\D', '')
# keep only first 2 characters of age
df['Age'] = df['Age'].str[:2]
```

```python
# split Preference of Trek Days (Select all that apply)	 into a list SEPERATED BY ; value should be a list
df['Preference of Trek Days (Select all that apply)'] = df['Preference of Trek Days (Select all that apply)']
# replace all non a-z characters with space in Languages (Separated by space)	
df['Languages (Separated by space)'] = df['Languages (Separated by space)'].str.replace(r'[^a-zA-Z]', ' ')


# REMOVE Unnamed: 18 column
# remove Username column
# remove Name column
# remove Timestamp column
df = df.drop(['Username'], axis=1)
df = df.drop(['Unnamed: 18'], axis=1)
df = df.drop(['Name'], axis=1)
df = df.drop(['Timestamp'], axis=1)
```

```python
df
```

```python
# turn all string columns into lower case
df = df.apply(lambda x: x.astype(str).str.lower())
```

```python
df
```

```python
# make sure height is not more than 3 digits
df['Height (in cm)'] = df['Height (in cm)'].str[:3]
# parse height to int
df['Height (in cm)'] = df['Height (in cm)'].astype(int)
# make sure weight is not more than 500kg
# parse the string to int and then convert it back to string
df['Weight (in kg)'] = df['Weight (in kg)'].apply(lambda x: str(int(x) if int(x) < 500 else x[:2]))
# make sure age is not more than 99
df['Age'] = df['Age'].apply(lambda x: str(int(x) if int(x) < 99 else 99))
# # number of treks should be 2 digits or less
# # if number of digits is more than 4, then keep only the first 2 digits
# df['Number of previous treks'] = df['Number of previous treks'].apply(lambda x: str((x) if len(x) == 4 else x[:2]))
# # if number of digits is 3, then keep only the last 2 digits
# df['Number of previous treks'] = df['Number of previous treks'].apply(lambda x: str((x) if len(x) == 3 else x[-2:]))
```

```python
df.to_json('../Data/googleform_cleaned.json', orient='records')
```
